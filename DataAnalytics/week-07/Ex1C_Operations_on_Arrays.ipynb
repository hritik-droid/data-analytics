## Sample Arrays

```
r_1 = np.random.randint(1, 20, size=2)

r_2 = np.random.randint(1, 20, size=(1, 2))

r_3 = np.random.rand(2, 1)

r_4 = np.random.randint(20, 100, size=(4, 5))

r_5 = np.random.rand(4, 5)

print(r_1)
print()
print(r_2)
print()
print(r_3)
print()
print(r_4)
print()
print(r_5)
```

## Permutations

```
print(r_4)
```

```
np.random.shuffle(r_4)

print(r_4)
```

The rows of the array were shuffled in place.
The values inside each row stayed together, but the order of the rows changed.

```
new_shuffle = np.random.permutation(r_4)

print("Permutation:")
print(new_shuffle)

print("\nOriginal r_4:")
print(r_4)
```

shuffle changes the original array directly.

permutation creates a shuffled copy while leaving the original array unchanged.

## Mathematical Operations

```
print(r_1)
print(r_1.shape)

print()

print(r_2)
print(r_2.shape)

result = np.add(r_1, r_2)

print()
print(result)
print(result.shape)
```

```
print(r_1)
print(r_1.shape)

print()

print(r_3)
print(r_3.shape)

result = np.add(r_1, r_3)

print()
print(result)
print(result.shape)
```

```
print(r_2)
print(r_2.shape)

print()

print(r_4)
print(r_4.shape)

result = np.add(r_2, r_4)

print()
print(result)
print(result.shape)
```

```
try:
    print(np.add(r_1, r_4))
except ValueError as e:
    print(e)

# r_1 cannot be added to r_4 because their shapes are not compatible for broadcasting.
```

```
print(np.add(np.array([1]), r_4))
```

```
print(np.add(np.array([[1]]), r_4))
```

```
print(np.add(np.array([1, 1, 1, 1, 1]), r_4))
```

```
try:
    print(np.add(np.array([[1], [1]]), r_4))
except ValueError as e:
    print(e)

# This produces a broadcasting error because the shapes are incompatible.
```

Broadcasting allows NumPy to combine arrays with different shapes when their dimensions are compatible.

Smaller arrays are stretched across larger arrays so arithmetic can happen element by element.

The resulting shape is determined by the larger compatible array.

## Grocery Problem

```
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print("Bag 1:", bag1)
print("Bag 2:", bag2)
```

```
print(f"The first bag contains {np.sum(bag1)} items, and the second bag contains {np.sum(bag2)} items.")
```

```
totals = bag1 + bag2

print(f"There are {totals[0]} vegetables, {totals[1]} fruits, {totals[2]} boxes of snacks, and {totals[3]} beverages in the two grocery bags.")
```

## Teamwork

```
ali = np.array([5, 4, 6, 4, 2])

print(ali)
```

```
ben = np.array([0, 5, 5, 3, 4])

print(ben)
```

```
carla = np.array([3, 6, 0, 3, 4])

print(carla)
```

```
all_knocks = np.array([ali, ben, carla])

print(all_knocks)
```

```
person_totals = np.sum(all_knocks, axis=1)

group_total = np.sum(all_knocks)

print("Ali total:", person_totals[0])
print("Ben total:", person_totals[1])
print("Carla total:", person_totals[2])
print("Group total:", group_total)
```

```
daily_totals = np.sum(all_knocks, axis=0)

rolling_totals = np.cumsum(daily_totals)

estimated_votes = rolling_totals * 0.5

print("Estimated rolling votes by day:")
print(estimated_votes)
```
