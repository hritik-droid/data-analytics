# Exercise 2.C Working with multiple DataFrames

```
import numpy as np
import pandas as pd
```

## Basic Aggregations

```
data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North'],
    'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
    'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]
}

df_sales = pd.DataFrame(data)

df_sales
```

```
df_sales["Revenue"] = df_sales["Units_Sold"] * df_sales["Price"]

df_sales
```

```
total_units_sold = df_sales["Units_Sold"].sum()
total_revenue = df_sales["Revenue"].sum()

print("Total Units Sold:", total_units_sold)
print("Total Revenue:", total_revenue)
```

```
average_units_sold = df_sales["Units_Sold"].mean()
average_revenue = df_sales["Revenue"].mean()

print("Average Units Sold:", average_units_sold)
print("Average Revenue:", average_revenue)
```

## Grouping Data

```
group_product = df_sales.groupby("Product")[["Units_Sold", "Revenue"]].sum()

group_product
```

```
group_region = df_sales.groupby("Region")[["Units_Sold", "Revenue"]].mean()

group_region
```

```
group_product_region = df_sales.groupby(
    ["Product", "Region"]
)[["Units_Sold", "Revenue"]].sum()

group_product_region
```

## Sorting and Ranking Data

```
sorted_units = df_sales.sort_values(by="Units_Sold")

sorted_units
```

```
sorted_revenue = df_sales.sort_values(by="Revenue", ascending=False)

sorted_revenue
```

```
df_sales["Revenue_Rank"] = df_sales["Revenue"].rank(ascending=False)

df_sales
```

## Merging DataFrames

```
product_data = {
    'Product': ['A', 'B', 'C'],
    'Category': ['Electronics', 'Furniture', 'Appliances']
}

df_products = pd.DataFrame(product_data)

df_products
```

```
df_merged = pd.merge(df_sales, df_products, on="Product")

df_merged
```

```
category_revenue = df_merged.groupby("Category")["Revenue"].sum()

category_revenue
```
