# Exercise 2.B Cleaning up data

```
import numpy as np
import pandas as pd
```

```
# Make sure test_data.csv exists in the same folder before running this notebook.

df = pd.read_csv("test_data.csv")

df
```

```
df.describe()
```

## Filtering a DataFrame

```
age_filter = df[df["Age"] > 18]

age_filter
```

```
score_filter = df[(df["Score"] > 50) & (df["Score"] < 100)]

score_filter
```

## Modifying a DataFrame

```
df_cleaned = df[df["Score"] >= 0]

df_cleaned
```

```
df_cleaned = df_cleaned.drop(columns=["City"])

df_cleaned
```

```
mean_age = df_cleaned["Age"].mean()

df_cleaned["Age"] = df_cleaned["Age"].fillna(mean_age)

df_cleaned
```

```
mean_score = df_cleaned["Score"].mean()

df_cleaned["Score"] = df_cleaned["Score"].replace(0, mean_score)

df_cleaned
```

```
df_cleaned = df_cleaned.dropna()

df_cleaned
```

```
df_cleaned.index = df_cleaned["ID"]

df_cleaned = df_cleaned.drop(columns=["ID"])

df_cleaned
```
