# Exercise 3.A Getting familiar with Matplotlib

```
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

print(x)
print(y)
```

```
plt.plot(x, y)

plt.show()
```

```
plt.plot(x, y, marker='o')

plt.show()
```

```
plt.plot(x, y, marker='o')

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Styling Plots

```
plt.plot(x, y, linestyle='--', color='red')

plt.title('Dashed Red Line')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```
plt.plot(
    x,
    y,
    marker='s',
    markersize=10,
    markerfacecolor='green'
)

plt.title('Square Markers')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```
plt.plot(x, y, marker='^', color='blue')

plt.title('Triangle Marker')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()

plt.plot(x, y, marker='*', color='purple')

plt.title('Star Marker')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```
