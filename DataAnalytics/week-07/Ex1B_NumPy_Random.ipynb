# Exercise 1.B NumPy Random

```
import numpy as np
```

## Generating Random Numbers

```
r_1 = np.random.randint(1, 20, size=2)

print(r_1)
```

```
r_2 = np.random.randint(1, 20, size=(1, 2))

print(r_2)
```

```
r_3 = np.random.rand(2, 1)

print(r_3)
```

```
r_4 = np.random.randint(20, 100, size=(4, 5))

print(r_4)
```

```
r_5 = np.random.rand(4, 5)

print(r_5)
```

## Generating Random Choices from an Array

```
students = np.array([
    "Alex", "Jordan", "Taylor", "Morgan",
    "Casey", "Riley", "Jamie", "Avery",
    "Quinn", "Cameron", "Drew", "Skyler"
])

print(students)
```

```
call_on = np.random.choice(students)

print(call_on)
```

```
breakout_groups = np.random.choice(students, size=(3, 4))

print(breakout_groups)

# This is not the best way to assign breakout groups because students may be repeated.
# Some students could appear more than once, while others may not appear at all.
```

```
call_on_weighted = np.random.choice(
    students,
    size=5,
    p=[0.30, 0.10, 0.10, 0.10, 0.10, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.00]
)

print(call_on_weighted)

# Probability explanation:
# Alex has a 30% chance of being selected.
# Skyler has a 0% chance of being selected.
# Jordan, Taylor, Morgan, Casey, and Riley each have a 10% chance.
# Jamie, Avery, Quinn, Cameron, and Drew each have a 5% chance.
# The probabilities add up to 100%.
```

## Generating a Normal Distribution

```
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))

print(bell_curve)
```
