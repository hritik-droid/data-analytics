```
import numpy as np
import pandas as pd
```

## Series

```
aa = [1, 2, 3, 4]

ab = (5, 6, 7, 8)

ac = {9, 10, 11, 12}

ad = {
    'first': 13,
    'second': 14,
    'third': 15,
    'fourth': 16
}
```

```
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac))
ser4 = pd.Series(ad)

print(ser1)
print()
print(ser2)
print()
print(ser3)
print()
print(ser4)
```

```
# This will produce a TypeError because Series expects one iterable object.

try:
    ser5 = pd.Series(17, 18, 19, 20)
except TypeError as e:
    print(e)
```

## DataFrames from Series

```
df1 = pd.DataFrame(ser1)

df1
```

```
df1.insert(1, "Column2", ser2)
df1.insert(2, "Column3", ser3)

df1
```

```
df1["Column4"] = ser4

df1

# ser4 uses dictionary keys as index labels instead of numbered indexes.
# Because the indexes do not match the DataFrame row indexes,
# pandas fills the unmatched rows with NaN values.
```

```
df1.loc[4] = [5, 9, 12, 0]

df1
```
