# Exercise 1.A NumPy: Manipulating Arrays

```
import numpy as np
```

## Looping Through Arrays

```
arrA = np.array([
    [11, 12, 13],
    [14, 15, 16]
])

print(arrA)
```

```
for x in arrA:
    print(x * 2)
```

```
for x in arrA:
    for y in x:
        print(y * 2)
```

The first loop goes through each row of the array and multiplies the entire row by 2.

The second loop goes through each individual value inside the rows and multiplies each number separately.

```
for x in arrA:
    loop_result = x * 2
    print(loop_result)

# This result is different because x represents an entire row,
# so multiplying x by 2 multiplies every element in that row.
```

## Joining and Stacking Arrays

```
joined_array_1 = np.concatenate((arrA, arrA))

print(joined_array_1)
```

```
joined_array_2 = np.concatenate((arrA, arrA), axis=1)

print(joined_array_2)
```

```
stacked_array_1 = np.stack((joined_array_1, joined_array_1))

print(stacked_array_1)
```

```
# This cell demonstrates the error caused by mismatched array shapes

try:
    stacked_array_2 = np.stack((joined_array_1, joined_array_2))
    print(stacked_array_2)
except ValueError as e:
    print(e)

# The arrays cannot be stacked because they do not have the same shape.
```

```
another_joined_array = np.concatenate((joined_array_1, arrA))

print(another_joined_array)
```

```
# This cell demonstrates the error caused by mismatched row counts

try:
    another_2 = np.concatenate((joined_array_1, arrA), axis=1)
    print(another_2)
except ValueError as e:
    print(e)

# The arrays cannot be concatenated along axis 1 because the number of rows does not match.
```

```
h_stack = np.hstack((arrA, arrA))
# hstack joins arrays horizontally (side by side)

v_stack = np.vstack((arrA, arrA))
# vstack joins arrays vertically (top and bottom)

d_stack = np.dstack((arrA, arrA))
# dstack stacks arrays depth-wise into a third dimension

print("Horizontal Stack:")
print(h_stack)

print("\nVertical Stack:")
print(v_stack)

print("\nDepth Stack:")
print(d_stack)
```

## Splitting Arrays

```
splitsky = np.array_split(joined_array_1, 2)

print(splitsky)

# Result:
# This creates 2 arrays.
# Each resulting array is still 2-dimensional.
```

```
spliterino = np.array_split(joined_array_1, 3)

print(spliterino)

# Result:
# This creates 3 arrays.
# The arrays are split as evenly as possible and remain 2-dimensional.
```

## Searching and Replacing Array Values

```
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')

print(result_1)
```

## Sorting and Filtering Arrays

```
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

arrB_sorted = np.sort(arrB)

print(arrB_sorted)
```

```
print(arrB)

z = arrB < 7

arrC = arrB[z]

print(arrC)
```

```
y = arrB >= 7

arrD = arrB[y]

print(arrD)
```
