# Exercise 5.A NumPy Fundamentals

This notebook practices NumPy basics, including creating arrays, checking data types, transposing arrays, slicing/indexing, and creating multidimensional arrays.

```
import numpy as np
```

## Array Basics

```
# Create a standard Python list
fruits = ["apple", "banana", "cherry", "date"]
fruits
```

```
# Create a one-dimensional NumPy array from the fruits list
arr_1 = np.array(fruits)
print(arr_1)
```

```
# Check the data type of the values in arr_1
print(arr_1.dtype)

# The dtype returned is usually <U6.
# The U means Unicode string.
# The number means the maximum string length in the array.
```

```
# Create a two-dimensional NumPy array
arr_2 = np.array([
    [10, 11, 12],
    [13, 14, 15]
])

print(arr_2)
```

```
# Print arr_2 with its dimensions transposed
print(arr_2.T)
```

```
# Check whether arr_2 was permanently changed
print(arr_2)

# arr_2.T does not permanently alter arr_2.
# It returns a transposed view/output, but arr_2 itself stays the same.
```

```
# Create another two-dimensional NumPy array
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

print(arr_3)
```

```
# Create arr_4 using selected values from arr_3
# arr_3[0, 2] is 5
# arr_3[1, 1] is 4
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])

print(arr_4)
```

```
# Create a three-dimensional array containing all 1s
# Each dimension has two values
arr_5 = np.ones((2, 2, 2))

print(arr_5)
```
