# Exercise 2.A Using built-in functions
## Lab 1. Working with names – parsing strings

```
test_name = 'John Doe'
test_name
```

```
def parse_and_display(name):
    parts = name.split()
    
    if len(parts) == 1:
        first = parts[0]
        print(f"Full name: {name}")
        print(f"First name: {first}")
        
    elif len(parts) == 2:
        first, last = parts
        print(f"Full name: {name}")
        print(f"First name: {first}")
        print(f"Last name: {last}")
        
    elif len(parts) == 3:
        first, middle, last = parts
        print(f"Full name: {name}")
        print(f"First name: {first}")
        print(f"Middle name: {middle}")
        print(f"Last name: {last}")
```

```
parse_and_display(test_name)
```

```
test_list = ['Grace Flores', 'JB Oinonen', 'Ken Penn']
```

```
for person in test_list:
    parse_and_display(person)
    print()
```

```
parse_and_display('Hritik Zope')
```

```
parse_and_display('Lorde')
print()
parse_and_display('Billie Eilish')
print()
parse_and_display('Megan Thee Stallion')
```
