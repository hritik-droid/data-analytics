# Exercise 2.B Using library functions
## Lab 2. Working with datetime functions

```
import datetime
```

# Today

```
today = datetime.datetime.now()
today
```

```
print(today.strftime("Today is %A, %B %d, %Y"))
```

# My birthday

```
# Replace this date with your own birthday if needed
birthday = datetime.date(2000, 1, 1)
birthday
```

```
print(birthday.strftime("My birthday is %m/%d/%Y"))
```

# Next 90 Days

```
ninety_d = today + datetime.timedelta(days=90)
ninety_d
```

```
print(ninety_d.strftime("90 days from today is %B %d, %Y"))
```

# Time for Dinner

```
dinner_time = datetime.time(19, 30)
dinner_time
```

```
print(dinner_time.strftime("Let’s meet for dinner at %I:%M %p"))
```
